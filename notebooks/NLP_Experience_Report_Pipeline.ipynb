```python
from transformers import pipeline

# Load the distilled roberta emotion classifier pipeline
emotion_pipeline = pipeline(
    "sentiment-analysis",
    model="j-hartmann/emotion-english-distilroberta-base",
    top_k=7
)

def predict_emotions(text, top_k=7):
    """Returns top K emotions from the given text."""
    try:
        text = text.strip()
        if not text:
            return "Error: Text is empty after stripping"

        # Truncate long text (max token limit for the model)
        result = emotion_pipeline(text[:256])
        # Sort emotions by descending score
        emotions = sorted(result[0], key=lambda x: -x['score'])
        return emotions[:top_k]

    except Exception as e:
        return f"Error: {e}"
```

```python
import pandas as pd
import matplotlib.pyplot as plt

def visualize_emotions(emotions):
    if not emotions:
        print("No emotions to show.")
        return

    labels = [e["label"] for e in emotions]
    scores = [e["score"] for e in emotions]

    df = pd.DataFrame({"Emotion": labels, "Score": scores})
    df = df.sort_values("Score", ascending=True)

    # Plot
    plt.figure(figsize=(6, 4))
    bars = plt.bar(df["Emotion"], df["Score"], color='orange')
    plt.xlabel("Emotions")
    plt.ylabel("Score")
    plt.title("Emotions Detected in Experiences")

    # Add value labels above bars
    for bar in bars:
        height = bar.get_height()
        plt.text(
            bar.get_x() + bar.get_width() / 2,  # x-position (center of bar)
            height,                             # y-position (top of bar)
            f"{height:.2f}",                    # format to 2 decimals
            ha="center", va="bottom"            # align center, slightly above
        )

    plt.show()
```

```python
from sentence_transformers import SentenceTransformer
import numpy as np

model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
embeddings = model.encode(trip_reports)
print(f"Shape of embeddings: {embeddings.shape}")
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity_matrix = cosine_similarity(embeddings)
print("Cosine similarity matrix:\n", similarity_matrix)
```

```python
def most_similar_reports(sim_matrix, top_k=4):
    for i in range(sim_matrix.shape[0]):
        # Get indices sorted by similarity score (excluding self)
        similar_idx = np.argsort(-sim_matrix[i])
        similar_idx = similar_idx[similar_idx != i]  # remove self
        top_similar = similar_idx[:top_k]
        print(f"\nReport Similarity to Report {i+1}:\n")
        for idx in top_similar:
            print(f" - Report {idx+1} ({sim_matrix[i][idx]:.3f})")

most_similar_reports(similarity_matrix)
```

```python
def similarity_label(score):
    if score >= 0.75:
        return "very similar"
    elif score >= 0.70:
        return "similar"
    elif score >= 0.65:
        return "somewhat similar"
    elif score >= 0.50:
        return "not similar"
    else:
        return "not at all similar"

def most_similar_reports_labeled(sim_matrix, top_k=4):
    for i in range(sim_matrix.shape[0]):
        similar_idx = np.argsort(-sim_matrix[i])
        similar_idx = similar_idx[similar_idx != i]
        top_similar = similar_idx[:top_k]
        print(f"\nTop {top_k} reports similar to Report {i+1}:")
        for idx in top_similar:
            label = similarity_label(sim_matrix[i][idx])
            print(f" - Report {idx+1} is {label} ({sim_matrix[i][idx]:.3f})")

most_similar_reports_labeled(similarity_matrix)
```
